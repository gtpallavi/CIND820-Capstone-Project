```python
#!pip install snscrape
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Creating list to append tweet data to
tweets_list2 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('COVID-19 lang:en -is:retweet since:2020-03-13 until:2022-03-14').get_items()):
    if i>14999:
        #print(type(tweet.hashtags))
        break
    tweets_list2.append([tweet.user.username, tweet.user.location, tweet.user.description, 
                         tweet.user.created, tweet.user.followersCount, 
                         tweet.user.friendsCount, tweet.user.favouritesCount, tweet.user.verified,
                         tweet.date, tweet.id, tweet.content, 
                         tweet.hashtags, tweet.sourceLabel, tweet.retweetedTweet])
#     user_details = sntwitter.TwitterUserScraper(tweet.user).entity
#     print(user_details.username)
tweets_list2    
    
# Creating a dataframe from the tweets list above
tweets_df2 = pd.DataFrame(tweets_list2, columns=['user_name', 'user_location', 'user_description', 
                                                 'user_created', 'user_followers', 
                                                 'user_friends', 'user_favourites', 'user_verified',
                                                 'date', 'tweetid', 'text', 'hashtags', 'source', 'is_retweet'])
tweets_df2.to_csv('March14FullTweets.csv', encoding = 'utf-8-sig', index = False)
tweets_df2
```
